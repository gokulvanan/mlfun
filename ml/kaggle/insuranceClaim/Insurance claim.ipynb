```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_file='train.csv'
train_data = pd.read_csv(train_file)
train_data.head()
```

```python
train_data.describe()
```

```python
def scatter_plot(x, y):
    plt.figure()
    plt.title(x+" vs "+y)
    plt.legend()
    plt.scatter(train_data[x],train_data[y])
    plt.xlabel(x)
    plt.ylabel(y)
    plt.show()
```

```python
scatter_plot("cont14",'loss')
```
